```python
import numpy as np
from ruptures_apart import apart
import matplotlib.pyplot as plt
import time
pi = np.pi
np.random.seed(1234)
```

```python
signal_lengths = [10**i for i in range(9)]
execution_times = []

for n in signal_lengths:
    signal = (np.random.random(size=(n, 3)) * (2 * pi - 1e-10)) + 1e-10
    start_time = time.time()
    chpnts = apart(signal, pen=1, nStates=1)
    end_time = time.time()
    exetime = end_time - start_time
    execution_times.append(exetime)
```

```python
# Plotting
plt.figure(figsize=(8, 3))
plt.plot(np.log10(signal_lengths[1:]), np.log10(execution_times[1:]), marker='o', color='black')
plt.xlabel('log signal_length')
plt.ylabel('log exe_time')
plt.title('Execution Time vs Signal Length')
plt.grid(True)
plt.savefig("figs/runtime_analysis.png", bbox_inches='tight')
plt.show()
```
